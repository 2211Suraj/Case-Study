```python
def matix(data):
    relation = data.corr().round(2)
    valuemask = np.zeros_like(relation, dtype=np.bool)
    valuemask[np.triu_indices_from(valuemask)] = True
    f, ax = plt.subplots(figsize=(22, 15))
    color_map = sns.diverging_palette(220, 10, as_cmap=True)
    correlation=sns.heatmap(relation, mask=valuemask, cmap=color_map, vmin=-1, vmax=1, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)
    plt.tight_layout()
    return correlation
```

```python
matix(df) #correlation plots 
```

#Machine learning models

#Splitting dataset into traning and testing 

```python
training_set_x = df.drop(['fetal_health'], axis  = 1)
testing_set_y = df['fetal_health']
```

#Using train test split 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(training_set_x,testing_set_y,test_size = 0.25, random_state = 0)
```

#Training and testing data on model

#XgBoost classifier

```python
import xgboost as xgb
boost_classifier= xgb.XGBClassifier()
```

```python
boost_classifier.fit(x_train, y_train)
```

```python
prediction= boost_classifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print("Accuracy score:-", accuracy_score(prediction, y_test))
```

#Prediction and evalution of dataset

```python
print("Classification_report:-\n", classification_report(prediction, y_test))
```

```python
print("confusion_matrix:-\n", confusion_matrix(prediction, y_test))
```
